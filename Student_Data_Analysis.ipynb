### - Ranking Logic for Top Performers

Since multiple students scored the same final marks, additional criteria were used
to determine ranking. Attendance percentage, improvement score, and daily study hours
were considered as tie-breaking factors to rank students fairly.


```python
df['pass_fail'] = df['pass_fail'].astype('string')
df['pass_fail'].dtypes
df['pass_fail'] = df['pass_fail'].replace({'1':'Pass', '0': 'Fail'})
df['pass_fail'] = df['final_score'].apply(
    lambda x: 'Pass' if x >= 40 else 'Fail'
)
df['grade'] = df['final_score'].apply(lambda x: 'A' if x >= 85 else ('B' if x >= 70 else ('C' if x >= 55 else ('D' if x >= 40 else 'F'))))

df.head(20)
```

# The pass/fail status was recalculated based on final scores to ensure correct classification.


```python
def result_status(marks):
    if marks < 40:
        return "Fail"
    elif marks <= 45:
        return "Just Passed"
    else:
        return "Pass"

df['result'] = df['final_score'].apply(result_status)
df['result'].value_counts()
df.head(20)
```

## Performance Categorization Summary

- Students were categorized into Pass, Fail, and Just Passed groups based on their marks.
This classification helps identify students who require academic intervention and
those who are performing well.
- Grade recalculate according to its marks


```python
HIGH_STUDY = df['daily_study_hours'].quantile(0.75) # it store the 75th percentile value
LOW_SCORE = 40

high_efforts_low_score = df[(df['daily_study_hours'] >= HIGH_STUDY) & (df['final_score'] < LOW_SCORE)]
high_efforts_low_score[['student_id', 'daily_study_hours', 'final_score']]


```
